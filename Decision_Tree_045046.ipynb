**3.2. Data Analysis**

**3.2.1.1. Decision Tree (Base Model):**

- **Overview:** Decision trees are hierarchical structures that recursively partition the feature space into regions, making decisions based on the values of input features.
- **Methodology:** The decision tree algorithm used in this analysis employs a recursive binary splitting approach based on the Gini impurity criterion.
- **Evaluation Metrics:**
  - **Accuracy:** The primary metric used to assess the decision tree's performance in correctly classifying instances.
  - **Training Time:** The time taken by the model to learn from the training data and construct the decision tree structure.
  - **Memory Usage:** The amount of memory required to store the decision tree model and associated data structures.

**3.2.1.2. Logistic Regression:**

- **Overview:** Logistic regression is a linear classification algorithm that models the probability of the outcome using the logistic function.
- **Methodology:** The logistic regression algorithm estimates coefficients for each feature to fit a linear decision boundary that separates different classes.
- **Evaluation Metrics:**
  - **Accuracy:** Measures the proportion of correctly classified instances by the logistic regression model.
  - **Training Time:** The time taken to train the logistic regression model using the training data.
  - **Memory Usage:** The memory footprint of the logistic regression model, including coefficients and intercept.

**3.2.1.3. Support Vector Machine (SVM):**
- **Overview:** SVM is a discriminative classification algorithm that finds the optimal hyperplane to separate different classes by maximizing the margin between them.
- **Methodology:** SVM constructs a hyperplane in the feature space that maximizes the margin between classes while minimizing classification errors.
- **Evaluation Metrics:**
  - **Accuracy:** Measures the proportion of correctly classified instances by the SVM model.
  - **Training Time:** The time taken to train the SVM model and identify the optimal hyperplane.
  - **Memory Usage:** The memory required to store support vectors and associated parameters of the SVM model.

**3.2.1.4. K Nearest Neighbour (KNN):**

- **Overview:** KNN is a non-parametric classification algorithm that classifies a new data point based on the majority class among its k nearest neighbors.
- **Methodology:** KNN assigns the class label of the majority of its k nearest neighbors to the new data point, using a distance metric such as Euclidean distance.
- **Evaluation Metrics:**
  - **Accuracy:** Measures the proportion of correctly classified instances by the KNN model.
  - **Training Time:** KNN is an instance-based learning algorithm and does not involve explicit training. However, the time taken to classify new instances can be considered as a measure of efficiency.
  - **Memory Usage:** The memory required to store the entire training dataset for efficient retrieval during classification.

By evaluating these classification algorithms based on accuracy, training time, and memory usage, we can determine the most suitable model for our specific dataset and application requirements.

**Analysis:**

**Accuracy:** It is one of the most intuitive performance measures and it is simply a ratio of correctly predicted observations to the total observations. In other terms, it is defined as the number of correct predictions made by the model divided by the total number of predictions.

Accuracy (0.48 or 48%):
This indicates that the decision tree correctly predicts the outcome 48% of the time across all classes. This is a moderate level of accuracy, suggesting that nearly half of the predictions made by the model are correct. An accuracy of 48% is relatively low for many applications, indicating that the model might not be very effective or that it could be struggling with the complexity or diversity of the dataset.


**Recall (or Sensitivity):** This measures how well the model is able to identify the actual positives for each class. It's particularly important when the cost of false negatives is high.

Recall (0.43 or 43%):
Recall, or the true positive rate, measures the model's ability to correctly identify actual positives. A recall of 43% means that the model identifies 43% of all relevant cases within the dataset. This is relatively low, indicating that more than half of the positive cases are missed by the model.

**Precision:** This assesses the accuracy of the positive predictions. High precision means that an algorithm returned substantially more relevant results than irrelevant ones.

Precision (0.53 or 53%):
Precision assesses the accuracy of the positive predictions made by the model. A precision of 53% suggests that when the model predicts an instance to belong to a particular class, that prediction is correct about 53% of the time. This is moderate but suggests that almost half of the positive predictions made by the model are false positives.

**F1-Score:** Combines recall and precision into a single metric by taking their harmonic mean. It balances both the recall and precision, which is especially useful when you have an uneven class distribution.

F1-Score (0.41 or 41%):
The F1-score is a harmonic mean of precision and recall and is particularly useful when the classes are imbalanced. An F1-score of 41% is quite low, indicating that the balance between precision and recall is not optimal. This low F1-score reinforces the notion that the model is not performing well in either correctly identifying all relevant instances or in minimizing the number of incorrect positive predictions.

**Overall Interpretation:**

The combination of these metrics suggests that the decision tree model is performing suboptimally in classifying the instances in your dataset. The low accuracy and F1-score indicate that the model struggles to effectively balance correct predictions and error minimization, which could be due to various factors such as overfitting, underfitting, not handling class imbalance effectively, or possibly requiring better feature selection or engineering.
The relatively low recall indicates that many actual positive cases are being missed, which can be critical depending on the application (like in medical diagnosis or fraud detection).
The moderate precision indicates that the model, while not extremely unreliable in its positive predictions, still incorrectly labels a significant number of instances as positive.

##**5. Managerial Insights**

**5.1. Appropriate Model: Compare and Contrast {Decision Tree | Logistic Regression | Support Vector Machine | K Nearest Neighbour}**

When choosing the most appropriate classification model, several factors must be considered, including model performance, computational efficiency, and interpretability. Let's delve deeper into the comparison and contrast of the Decision Tree, Logistic Regression, Support Vector Machine (SVM), and K Nearest Neighbour (KNN) models based on these aspects:

- **Decision Tree:**
  - Advantages:
    - Intuitive and easy to understand, making it suitable for stakeholders without advanced technical knowledge.
    - Can handle both numerical and categorical data effectively without extensive preprocessing.
    - Offers insights into feature importance and decision-making processes through visual representations of the tree structure.
  - Disadvantages:
    - Prone to overfitting, especially with deep trees, which can lead to poor generalization on unseen data.
    - Sensitive to small variations in the training data, resulting in different tree structures for slightly different datasets.
    - May struggle with capturing complex relationships between features due to its hierarchical nature.

- **Logistic Regression:**
  - Advantages:
    - Provides interpretable coefficients for each feature, allowing stakeholders to understand the impact of each variable on the outcome.
    - Robust to noise and outliers, making it suitable for datasets with noisy or irregularly distributed data.
    - Performs well when the relationship between features and the target variable is approximately linear.
  - Disadvantages:
    - Assumes a linear relationship between features and the log-odds of the outcome, which may not hold true for complex datasets.
    - Limited in capturing nonlinear relationships without feature engineering or polynomial transformations.
    - Susceptible to multicollinearity, where highly correlated features can lead to unstable coefficient estimates.

- **Support Vector Machine (SVM):**
  - Advantages:
    - Effective in high-dimensional spaces, making it suitable for datasets with many features.
    - Versatile with various kernel functions to capture nonlinear relationships between features.
    - Robust to overfitting, especially in high-dimensional spaces, due to the margin maximization objective.
  - Disadvantages:
    - Computationally intensive, especially with large datasets, as it involves solving a quadratic optimization problem.
    - Requires careful selection of hyperparameters and kernel functions, which may require extensive tuning.
    - Less interpretable compared to Decision Trees and Logistic Regression, as it lacks direct insights into feature importance and decision rules.

- **K Nearest Neighbour (KNN):**
  - Advantages:
    - Simple and intuitive, making it easy to understand and implement, particularly for baseline models.
    - Nonparametric nature allows flexibility in capturing complex decision boundaries without making strong assumptions about the data distribution.
    - Robust to noisy data and outliers, as it relies on local information from neighboring instances.
  - Disadvantages:
    - Computationally expensive during inference, especially with large datasets, as it requires calculating distances to all training instances.
    - Sensitive to the choice of the number of neighbors (k) and distance metric, which can significantly impact performance.
    - Memory-intensive, as it requires storing all training instances, making it impractical for datasets with a large number of observations.

```python
import pandas as pd
import time
import psutil
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Function to measure memory usage
def memory_usage():
    process = psutil.Process()
    return process.memory_info().rss / 1024 ** 2  # Memory usage in MB

# Data preprocessing and splitting
# Assuming you have your data loaded into product_inputs and product_output
product_inputs_train,product_inputs_test, product_output_train, product_output_test = train_test_split(product_inputs, product_output, test_size=0.2, random_state=42)

# Results storage
results = []

# Decision Tree
dt_start_time = time.time()
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(product_inputs_train, product_output_train)
dt_training_time = time.time() - dt_start_time
dt_memory_used = memory_usage()
dt_pred = dt_model.predict(product_inputs_test)
dt_accuracy = accuracy_score(product_output_test, dt_pred)
results.append({'Model': 'Decision Tree', 'Time Taken (s)': dt_training_time, 'Memory Used (MB)': dt_memory_used, 'Accuracy': dt_accuracy})

# KNN
k_values = [5, 7, 9, 11, 13, 15]
for k in k_values:
    knn_start_time = time.time()
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(product_inputs_train, product_output_train)
    knn_training_time = time.time() - knn_start_time
    knn_memory_used = memory_usage()
    knn_pred = knn.predict(product_inputs_test)
    knn_accuracy = accuracy_score(product_output_test, knn_pred)
    results.append({'Model': f'KNN (k={k})', 'Time Taken (s)': knn_training_time, 'Memory Used (MB)': knn_memory_used, 'Accuracy': knn_accuracy})

# Logistic Regression
lr_start_time = time.time()
logreg = LogisticRegression(random_state=45011, solver='liblinear')
logreg.fit(product_inputs_train, product_output_train)
lr_training_time = time.time() - lr_start_time
lr_memory_used = memory_usage()
lr_pred = logreg.predict(product_inputs_test)
lr_accuracy = accuracy_score(product_output_test, lr_pred)
results.append({'Model': 'Logistic Regression', 'Time Taken (s)': lr_training_time, 'Memory Used (MB)': lr_memory_used, 'Accuracy': lr_accuracy})

# SVM
svm_start_time = time.time()
svm_classifier = SVC(kernel='rbf', random_state=45011)
svm_classifier.fit(product_inputs_train, product_output_train)
svm_training_time = time.time() - svm_start_time
svm_memory_used = memory_usage()
svm_pred = svm_classifier.predict(product_inputs_test)
svm_accuracy = accuracy_score(product_output_test, svm_pred)
results.append({'Model': 'SVM', 'Time Taken (s)': svm_training_time, 'Memory Used (MB)': svm_memory_used, 'Accuracy': svm_accuracy})

# Create DataFrame
results_df = pd.DataFrame(results)

# Display the results
print(results_df)
```
